## RGB

Individual Red, Green and Blue channels which are non-perceptual and have strongly correlated channels. Not ideal for image processing or computer vision tasks but good enough for most applications.


```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load image
img = cv.imread('rgb-test.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Display image and individual channels
plt.figure(figsize=(16, 9))
plt.subplot(221)
plt.imshow(img)

plt.subplot(222)
plt.imshow(img[:,:,0], cmap='Reds')

plt.subplot(223)
plt.imshow(img[:,:,1], cmap='Greens')

plt.subplot(224)
plt.imshow(img[:,:,2], cmap='Blues')

plt.show()
```

## HSV

Hue, Saturation and Value channels. 

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load image
img = cv.imread('bbc_cat.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img, cv.COLOR_BGR2HSV)

# Display image and individual channels
plt.figure(figsize=(16, 9))
plt.subplot(221)
plt.imshow(img)

plt.subplot(222)
plt.imshow(img2[:,:,0], cmap='hsv')

plt.subplot(223)
plt.imshow(img2[:,:,1], cmap='Reds')

plt.subplot(224)
plt.imshow(img2[:,:,2], cmap='Greys')

plt.show()
```
