## Pin-hole Camera Model

A 3D point X is projected to an image point x (both expressed as homogenous coordinates). Lambda is the inverse depth of the 3D point. P is the projection matrix aka camera matrix.



The camera matrix (P) can be decomposed into:




```python
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt

class Camera(object):
    """ Class for representing pin-hole cameras. """

    def __init__(self, P):
        """ Initiliase P = K[R|t] camera model. """

        self.P = P
        self.K = None # Calibration matrix
        self.R = None # Rotation matrix
        self.t = None # Translation vector
        self.c = None # Camera Optical Centre
      
    def project(self, X):
        """ Project points in X (4*n Array) and normalise coordinates. """
        x = np.dot(self.P, X)

        for i in range(3):
            x[i] /= x[2]

        return x
  
def rotation_matrix(a):
    """ Creates a 3D rotation matrix for rotating around the axis of the vecotr a"""
    R = np.eye(4)
    # [X, Y, Z]
    R[:3, :3] = linalg.expm([[0, -a[2], a[1]], [a[2], 0, -a[0]], [-a[1], a[0], 0]])

    return R
        
```

```python
# load points
points = np.loadtxt('house.p3d').T
points = np.vstack((points, np.ones(points.shape[1])))

# setup camera
# np.eye creates a matrix with 1s diagonally and 0s elsewhere
# P is the projection matrix or camera matrix
P = np.hstack((np.eye(3), np.array([[0], [0], [-10]])))
cam = Camera(P)
x = cam.project(points)

# plot projection
plt.figure(figsize=(16, 9))

plt.subplot(221)
plt.plot(x[0], x[1], 'k.')
plt.title('Default conditions')

# Rotate on X-Axis
plt.subplot(222)
r = [0.01, 0, 0]
rot = rotation_matrix(r)

for t in range(20):
    cam.P = np.dot(cam.P, rot)
    x = cam.project(points)
    plt.plot(x[0], x[1], 'k.')
plt.title('Rotated on X-Axis')

plt.subplot(223)
r = [0, 0.01, 0]
rot = rotation_matrix(r)

for t in range(20):
    cam.P = np.dot(cam.P, rot)
    x = cam.project(points)
    plt.plot(x[0], x[1], 'k.')
plt.title('Rotated on Y-Axis')

plt.subplot(224)
r = [0, 0, 0.01]
rot = rotation_matrix(r)

for t in range(20):
    cam.P = np.dot(cam.P, rot)
    x = cam.project(points)
    plt.plot(x[0], x[1], 'k.')
plt.title('Rotated on Z-Axis')

plt.show()
```
