# Parametric (Global) Transformations

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def warpParametric(img, mat):
    img_new = np.zeros((img.shape[0]*mat[0, 0], img.shape[1]*mat[1, 1]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            coor = np.matmul(mat, [i, j])
            img_new[coor[0], coor[1]] = img[i, j]
            
    return img_new

img = cv.imread('bbc_cat.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')

plt.show()
```

## Scaling

Multiply coordinates with matrix below:

$
    \begin{bmatrix} 
        a & 0 \\
        0 & d \\
    \end{bmatrix}
$

```python
# Parametric transformations aka Global transformations
matrix = np.zeros((2, 2), dtype=np.int32)
matrix[0, 0] = 2
matrix[1, 1] = 1
print(matrix)

img_scaled = warpParametric(img, matrix)
plt.imshow(img_scaled, cmap='gray')
plt.show()
```
