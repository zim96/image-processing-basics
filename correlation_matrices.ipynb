```python
# Correlation matrices, what do they represent?
import numpy as np

X = np.random.randint(1, 10, size=(3,3))
XXT = X @ np.transpose(X)
XTX = np.transpose(X) @ X

print('X:')
print(X)
print('XT:')
print(np.transpose(X))
print('XXT:')
print(XXT)
print('XTX:')
print(XTX)
```

```python
from matplotlib import image
import matplotlib.pyplot as plt

def im2double(im, min_val, max_val)    :
    return (im.astype('float')-min_val)/(max_val-min_val)

img = im2double(image.imread('test_images/square.jpg'), 0, 255)
img = np.mean(img, -1)

plt.title('Image')
plt.imshow(img, cmap='gray')
plt.show()
```

```python
img_1 = img @ img.T
img_1 = im2double(img_1, np.min(img_1), np.max(img_1))
img_2 = img.T @ img
img_2 = im2double(img_2, np.min(img_2), np.max(img_2))

plt.subplot('121')
plt.title('AA`')
plt.imshow(img_1, cmap='gray')
plt.subplot('122')
plt.title('A`A')
plt.imshow(img_2, cmap='gray')
plt.show()
```
